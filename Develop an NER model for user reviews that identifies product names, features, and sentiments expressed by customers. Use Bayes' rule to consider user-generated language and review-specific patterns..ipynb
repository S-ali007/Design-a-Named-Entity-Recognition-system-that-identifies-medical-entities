```python
import pandas as pd

# Load the dataset
data = pd.read_csv("/content/test.ft.txt", sep="\t", header=None, names=["raw_data"])

# Extract sentiment and review_text from the raw_data column
data["sentiment"] = data["raw_data"].apply(lambda x: x.split()[0])
data["review_text"] = data["raw_data"].apply(lambda x: " ".join(x.split()[1:]))

# Preprocessing
data["review_text"] = data["review_text"].str.lower().str.replace('[^\w\s]', '').str.split()

# Display the first few rows of the preprocessed data
data.head()
```

```python
import spacy

# Download and install the pre-trained SpaCy model
!python -m spacy download en_core_web_sm

# Load the SpaCy model
nlp = spacy.load("en_core_web_sm")

# Annotate named entities
def annotate_ner(text):
    doc = nlp(" ".join(text))
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return entities

# Apply NER to each review
data["ner_entities"] = data["review_text"].apply(annotate_ner)

# Display the first few rows with NER annotations
data.head()
```
